# PREPROCESSING 
# Data has a lot none non-numerical values, to proceed further without losing information, this data has to be parsed to numeric. Also, some outliers should be dropped   

```python
student_depression = student_depression_original[student_depression_original['Profession'] =='Student']
```

```python
student_depression = student_depression.copy()

student_depression['suicidal_thoughts_present'] = (
    student_depression['Have you ever had suicidal thoughts ?']
    .astype(str)
    .map({'Yes': 1, 'No': 0})
)

student_depression['history_of_mental_illness'] = (
    student_depression['Family History of Mental Illness']
    .astype(str)
    .map({'Yes': 1, 'No': 0})
);
student_depression = student_depression.drop(
    columns=['Profession','Have you ever had suicidal thoughts ?','Family History of Mental Illness', 'id']
)
```

```python
student_depression = student_depression.reset_index().drop(columns=['index'])
```

### Rest of the non-numeric data can be encoded

```python
encoder = OneHotEncoder()
transformed_data_arr = encoder.fit_transform(student_depression[['Gender','City', 'Sleep Duration','Dietary Habits', 'Degree']])
transformed_data_df = pd.DataFrame(transformed_data_arr.toarray(), columns=encoder.get_feature_names_out())
transformed_data_df[transformed_data_df.isna().any(axis=1)]
transformed_data_df
```

```python
student_depression_ajasted = student_depression.drop(columns=['Gender','City','Sleep Duration','Dietary Habits','Degree'])
```

```python
student_depression_final= pd.concat([student_depression_ajasted, transformed_data_df], axis=1)
student_depression_final.head()
```

## Now, then all columns are numeric, we can check the correlation between the data and the target, which in this case is the Depression column

```python
mosr_corelated_values = np.abs((student_depression_final.corr()['Depression'])).sort_values(ascending=False).iloc[:10] #was 13 changed to 10 
mosr_corelated_values
```

# To reduce dimensionality, I'll take only columns with correlation present 

```python
mosr_corelated_values = mosr_corelated_values.index.to_list()
```

```python
student_depression_final= student_depression_final[mosr_corelated_values]
```

```python
student_depression_final.info() # this is the final frame 
```

```python
sns.heatmap(student_depression_final.corr())
```

# Choosing a model 

```python
X = student_depression_final.drop(columns=['Depression'])
y = student_depression_final['Depression']
```

### Splitting data into training and testing parts 

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
scaler = StandardScaler()
```

```python
X_train_s =  scaler.fit_transform(X_train)
X_test_s = scaler.fit_transform(X_test)
```

```python
forest =RandomForestClassifier()
```

```python
hyperparanms_forest= {'n_estimators':[100, 200, 500,1000], 'max_depth': [4,5,7,9,10,12], 'max_leaf_nodes': [8,10,12,14,15]}
```

```python
gridsearch_forest = GridSearchCV(forest, param_grid= hyperparanms_forest, cv=5, verbose=2)
```

```python
gridsearch_forest.fit(X_train_s, y_train)
```

```python
pd.DataFrame(gridsearch_forest.cv_results_).sort_values('rank_test_score')
```

```python
pipe_forest = Pipeline([
    ('poly_features', PolynomialFeatures(degree=2, interaction_only=False, include_bias=True)),
    ('scaler', StandardScaler()),
    ('forest_class', RandomForestClassifier(n_estimators=500, max_depth=4,max_leaf_nodes=15))
])
```

```python
result_forest = cross_validate(pipe_forest, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
```

```python
- np.mean(result_forest['test_score'])
```

# SVC let's try a different model

```python
svm = SVC(kernel='poly')
```

```python
grid_search_svm_params = {'C':[0.5,1,2,3,4,6,8,10], 'degree':[1,2,3,4]}
grid_svm = GridSearchCV(svm, param_grid=grid_search_svm_params, cv=5, verbose=2)
```

```python
grid_svm.fit(X_train_s, y_train)
```

```python
pd.DataFrame(grid_svm.cv_results_).sort_values('rank_test_score')
```

```python
svm_pipe= Pipeline([
    ('scaler', StandardScaler()),
    ('svm', SVC(kernel='poly', degree=1, C=6))
])
```

```python
result_svm = cross_validate(svm_pipe, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
```

```python
-np.mean(result_svm['test_score'])
```

# NNK model 

```python
nnk = KNeighborsClassifier()
```

```python
grid_params_nnk= {'n_neighbors':[20, 24, 25,26,28, 30,32]}
```

```python
grid_search_nnk = GridSearchCV(nnk, param_grid=grid_params_nnk, cv=5, verbose=2)
```

```python
grid_search_nnk.fit(X_train_s, y_train)
```

```python
pd.DataFrame(grid_search_nnk.cv_results_)
```

```python
pipe_nnk =Pipeline([
    ('scaler', StandardScaler()),
    ('nnk', KNeighborsClassifier(n_neighbors=28))
])
```

```python
results_nnk = cross_validate(pipe_nnk, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
```

```python
- np.mean(results_nnk['test_score'])
```

# SVC gave us a best results 

```python
svm_pipe.fit(X_train, y_train)
```

```python
y_predicted_cvm = svm_pipe.predict(X_test)
```

```python
accuracy= accuracy_score(y_test, y_predicted_cvm)
print('accuracy ', accuracy)
precision_score_svm = precision_score(y_test, y_predicted_cvm)
print('precision_score ', precision_score_svm)
recall_score_svm= recall_score(y_test, y_predicted_cvm)
print('recall_score ', recall_score_svm)
```

# Let's also double-check NNK 

```python
pipe_nnk.fit(X_train, y_train)
```

```python
y_nnk_predict = pipe_nnk.predict(X_test)
```

```python
accuracy_nnk= accuracy_score(y_test, y_nnk_predict)
print('accuracy ', accuracy_nnk)
precision_score_nnk = precision_score(y_test, y_nnk_predict)
print('precision_score ', precision_score_nnk)
recall_score_nnk= recall_score(y_test, y_nnk_predict)
print('recall_score ', recall_score_nnk)
```

# SVC  gives the best accuracy, therefore I'll choose this model 
## Let's fit all the data to the model now 

```python
svm_pipe.fit(X,y)
```
